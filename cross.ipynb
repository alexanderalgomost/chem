# Строим модель...

```python
# ошибка
def rmspe(y, a):
    # y=0 - игнорировать
    return np.mean((((y-a)/y)[y>0]) ** 2) ** .5

# ошибка
def prmse(y, a):
    # y=0 - игнорировать
    return (np.mean(((y-a)[y>0]) ** 2) ** .5)/np.mean(y[y>0])

# ошибка
def smape(y, a):
    # y=0 - игнорировать
    return np.mean((2*np.abs(y-a)/np.abs(y+a))[y>0])
```

```python
y = data['Sales'].values
```

Константные модели

```python
print rmspe(y, 0)
print rmspe(y, y[y>0].mean())
print rmspe(y, 1.1*y[y>0].mean())
print rmspe(y, 0.9*y[y>0].mean())
print rmspe(y, 0.65*y[y>0].mean())

print rmspe(y, np.median(y[y>0]))
print rmspe(y, 0.75*np.median(y[y>0]))
```

Медиана лучше среднего

Но если делать понижение проноза, то среднее лучше

```python
l = np.linspace(0,1,1001)
e = []
e2 = []
e3 = []
for x in l:
    a = x*y[y>0].mean() # np.median(y[y>0]) #
    e.append(rmspe(y, a))
    e2.append(prmse(y, a))
    e3.append(smape(y, a))

tmp = pd.DataFrame({'e': e, 'e2': e2, 'e3': e3})
tmp.index = l
ax = tmp.plot()
ax.set_title(u'Качество константного решения')
ax.set_xlabel(u'множитель')
ax.set_ylabel(u'ошибка')


print 'min rmspe = ' + str(min(e))
print 'min prmse = ' + str(min(e2))
print 'min smape = ' + str(min(e3))
```

**Вывод:** для нашего функционала хорош занижающий прогноз (в отличие от классических)
