```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.colors as colors
def plot3D(X):
    
    xs = [w[0] for w in X]
    ys = [w[1] for w in X]
    zs = [w[2] for w in X]

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(xs, ys, zs, marker='o')
    ax.set_xlabel('1st PC')
    ax.set_ylabel('2nd PC')
    ax.set_zlabel('3rd PC')
    plt.show()
    
def plot3D(X,labels):
    
    x = np.array([w[0] for w in X])
    y = np.array([w[1] for w in X])
    z = np.array([w[2] for w in X])
    
    t = colors.cnames.items()
    lab = np.column_stack((labels,labels,labels))
    fig = plt.figure()
    
    ax = fig.add_subplot(111, projection='3d')
    
    un = np.unique(labels)
    
    for lab in un:
        cond = labels == lab
        xs = x[cond]        
        ys = y[cond]
        zs = z[cond]        
        ax.scatter(xs, ys,zs, c=t[lab][0], marker='o')
        
from sklearn.decomposition import TruncatedSVD
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Normalizer
    
def plot2D(X, labels):
    
    svd = TruncatedSVD(2)
    lsa = make_pipeline(svd, Normalizer(copy=False))

    X_new = lsa.fit_transform(X)

    xs = [w[0] for w in X_new]
    ys = [w[1] for w in X_new]

    plt.figure()
    
    plt.scatter(xs,ys, c=labels, s=100, cmap=plt.cm.Spectral)
    xlabel('First principal component')
    ylabel('Second principal component')
    explained_variance = svd.explained_variance_ratio_.sum()
    plt.title("Explained variance of the SVD step: {}%".format(
    int(explained_variance * 100)))
    colorbar( values = range(len(np.unique(labels))))
    
    plt.show()
    
def plot3D_PC(X, labels):
    
    svd = TruncatedSVD(3)
    lsa = make_pipeline(svd, Normalizer(copy=False))

    t = colors.cnames.items()
    
    X_new = lsa.fit_transform(X)

    x = np.array([w[0] for w in X_new])
    y = np.array([w[1] for w in X_new])
    z = np.array([w[2] for w in X_new])
    
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    un = np.unique(labels)
    print 'unique', len(un)
    for lab in un:       
        cond = labels == lab
        xs = x[cond]        
        ys = y[cond]
        zs = z[cond]    
        ax.scatter(xs, ys,zs, c=t[lab][0], marker='o')
    
    explained_variance = svd.explained_variance_ratio_.sum()
    plt.title("Explained variance of the SVD step: {}%".format(
    int(explained_variance * 100)))
    
    plt.show()
```

```python
def to_nominal(data):
    d = {}
    i = 0
    un_mob =  data.unique()
    for cat in un_mob:
        d[cat] = i 
        i = i + 1
    res = np.zeros(data.shape, dtype = int)
    for w in d.keys():
        res[np.where(un_mob == w,)] = d[w]
    return [res,d]
```
